```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
diamonds = sns.load_dataset("diamonds")
```

```python
# Gráfico 1: Boxplot - Distribución de propinas por día
plt.figure()
sns.boxplot(x="day", y="tip", data=tips)
plt.title("Gráfico 1: Distribución de propinas por día")
plt.xlabel("Día de la semana")
plt.ylabel("Propina ($)")
plt.show()
```

```python
# Gráfico 2: Heatmap - Pasajeros por mes y año (flights)
pivot_flights = flights.pivot(index="month", columns="year", values="passengers")
plt.figure()
sns.heatmap(pivot_flights, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Gráfico 2: Pasajeros por mes y año")
plt.show()
```

```python
# Gráfico 3: Violin plot - Precio de diamantes por corte
plt.figure()
sns.violinplot(x="cut", y="price", data=diamonds.sample(1000), palette="muted")
plt.title("Gráfico 3: Precio de diamantes por tipo de corte")
plt.xlabel("Corte del diamante")
plt.ylabel("Precio ($)")
plt.show()
```

```python
# Gráfico 4: Count plot - Cantidad de fumadores por sexo
plt.figure()
sns.countplot(x="sex", hue="smoker", data=tips)
plt.title("Gráfico 4: Cantidad de fumadores por sexo")
plt.xlabel("Sexo")
plt.ylabel("Cantidad")
plt.show()
```
